```python
#Importing dependencies
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the CSV file (update the path if needed)
file_path = 'Resources/puntos-de-abastecimiento-de-agua-a-nivel-municipal-monterrey.csv'
df = pd.read_csv(file_path)
```

```python
print(df.columns)
```

```python
# Count occurrences of each water supply type in 'Nota'
supply_counts = df['Nota'].value_counts()
supply_counts.head(20)
```

```python
# Define the values based on your count data
supply_counts = {
    "Contenedor": 711 + 22 + 18 + 13 + 8 + 6 + 2 + 1,  # Sum of "1 Contenedor" and "2 Contenedores" or any other type of "contenedor"
    "Pipa": 210 + 196 + 98 + 88 + 14 + 7  # Sum of all other values
}
```

```python
# Plot the pie chart
plt.figure(figsize=(8, 8))
colors = ['yellow', 'purple']  # Contenedor (orange) & Pipa (blue)
plt.pie(supply_counts.values(), labels=supply_counts.keys(), autopct='%1.1f%%', startangle=90, colors=colors, wedgeprops={'edgecolor': 'black'})
plt.title('Distribution of Water Supply Type (Pipa vs Contenedor)')

# Save the pie chart as a PNG file before showing it
plt.savefig('Resources/pie_chart.png')

# Show the pie chart
plt.show()
```
